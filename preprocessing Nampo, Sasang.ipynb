##  전처리& 결과

```python
def preprocessing(df):
    df = df.apply(pd.to_numeric, errors = 'coerce')
    for elements in ['CO','CO2',"NO2",'PM2.5','PM10','HCHO']:
        print(elements)
        for years in [2015, 2016, 2017,2018,2019]:
            print(df[elements].loc[(df['datetime-year']==years) & (df[elements]>0)].mean()) # 평균
            print(df[elements].loc[(df['datetime-year']==years) & (df[elements]>0)].std())  # 표준편차
            print(df[elements].loc[(df['datetime-year']==years) & (df[elements]>0)].min())  # 최소값
            print(df[elements].loc[(df['datetime-year']==years) & (df[elements]>0)].max())  # 최대값
#         print()
preprocessing(nampo)
```

# 남포 OP/NON

```python
import pandas as pd

nampo = pd.read_csv('excel/남포_.csv', encoding = 'cp949')
nampo = nampo.loc[nampo['datetime-year']>=2015]

# nampo = nampo.loc[nampo['datetime-year']>=2015]
nampo_non = nampo.loc[((nampo['hour']<5))]
nampo_non
nampo_op = nampo.loc[((nampo['hour']>=5))]

nampo_op
```

```python
def preprocessing(df,df2):
    df = df.apply(pd.to_numeric, errors = 'coerce')
    df2 = df2.apply(pd.to_numeric, errors = 'coerce')
    
    for elements in ['CO','CO2',"NO2",'PM10','PM2.5']:
        for years in [2015, 2016, 2017,2018,2019]:
            print(df[elements].loc[(df['datetime-year']==years) & (df[elements]>0)].mean()) # 평균
            print(df[elements].loc[(df['datetime-year']==years) & (df[elements]>0)].std())  # 표준편차
            print(df2[elements].loc[(df2['datetime-year']==years) & (df2[elements]>0)].mean()) # 평균
            print(df2[elements].loc[(df2['datetime-year']==years) & (df2[elements]>0)].std())  # 표준편차
        print()
preprocessing(nampo_non,nampo_op)
```
