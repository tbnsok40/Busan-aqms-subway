```python
import pandas as pd
```

```python
def preprocessing(df):
    df = df.apply(pd.to_numeric, errors = 'coerce')
    
    for elements in ['CO','CO2',"NO2",'PM10','PM2.5','HCHO']:
#         print(elements)
        for years in [2015, 2016, 2017,2018,2019]:
            print(df[elements].loc[(df['year']==years) & (df[elements]>0)].mean()) # 평균
            print(df[elements].loc[(df['year']==years) & (df[elements]>0)].std())  # 표준편차
            print(df[elements].loc[(df['year']==years) & (df[elements]>0)].min())  # 최소값
            print(df[elements].loc[(df['year']==years) & (df[elements]>0)].max())  # 최대값
        print()
    
```

```python
s2w = pd.read_csv('excel/서면2w.csv', encoding = 'cp949')
s2w
```

```python
preprocessing(s2w)
```

```python
# 파일 바꿨음 hour column => 0 -> 24로 
# because, pandas doesn't cognize and(&) grammar.

s2w = pd.read_csv('excel/서면2w_.csv', encoding = 'cp949')
s2w = s2w.loc[s2w['year']>=2015]
# s2w = s2w.drop(['Unnamed: 0'], axis = 1)

# s1p = s1p.loc[(s1p['hour']>=5) & (s1p['hour']== 0)]

s2w_non = s2w.loc[((s2w['hour']<5))]
s2w_non
s2w_op = s2w.loc[((s2w['hour']>=5))]
s2w_op

```

```python
def preprocessing(df,df2):
    df = df.apply(pd.to_numeric, errors = 'coerce')
    df2 = df2.apply(pd.to_numeric, errors = 'coerce')
    
    for elements in ['CO','CO2',"NO2",'PM10','PM2.5']:
        for years in [2015, 2016, 2017,2018,2019]:
            print(df[elements].loc[(df['year']==years) & (df[elements]>0)].mean()) # 평균
            print(df[elements].loc[(df['year']==years) & (df[elements]>0)].std())  # 표준편차
            print(df2[elements].loc[(df2['year']==years) & (df2[elements]>0)].mean()) # 평균
            print(df2[elements].loc[(df2['year']==years) & (df2[elements]>0)].std())  # 표준편차
        print()
preprocessing(s2w_non,s2w_op)
```
