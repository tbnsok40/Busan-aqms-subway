```python
import pandas as pd
```

```python
def preprocessing(df):
    df = df.apply(pd.to_numeric, errors = 'coerce')
    
    for elements in ['CO','CO2',"NO2",'PM2.5','PM10']:
#         print(elements)
        for years in [2015, 2016, 2017,2018,2019]:
            print(df[elements].loc[(df['datetime-year']==years) & (df[elements]>0)].mean()) # 평균
            print(df[elements].loc[(df['datetime-year']==years) & (df[elements]>0)].std())  # 표준편차
            print(df[elements].loc[(df['datetime-year']==years) & (df[elements]>0)].min())  # 최소값
            print(df[elements].loc[(df['datetime-year']==years) & (df[elements]>0)].max())  # 최대값
        print()
    
```

```python
s1p = pd.read_csv('excel/서면1p.csv', encoding = 'cp949')
```

```python
s1p = s1p.loc[s1p['datetime-year']>=2015]
print(((((s1p['CO']==0)).sum())+(s1p['CO'].isnull().sum())) / 43824)
print(((((s1p['CO2']==0)).sum())+(s1p['CO2'].isnull().sum())) / 43824)
print(((((s1p['NO2']==0)).sum())+(s1p['NO2'].isnull().sum())) / 43824)
print(((((s1p['PM10']==0)).sum())+(s1p['PM10'].isnull().sum())) / 43824)
print(((((s1p['PM2.5']==0)).sum())+(s1p['PM2.5'].isnull().sum())) / 43824)
# print(((((s1p['HCHO']==0)).sum())+(s1p['HCHO'].isnull().sum())) / 43824)
```
