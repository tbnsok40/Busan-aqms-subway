# 광복

```python
import numpy as np
import pandas as pd
df = pd.read_csv('excel/AQMS_광복.csv')
df =df.loc[(df['year']<2020)&(df['year']>2014)]

df = df.apply(pd.to_numeric, errors = 'coerce')
df
```

```python
# df = df.loc[df['PM10']>0]
df['PM10'].isnull().sum()
```

```python
def solution(df):
    
    elements = ['CO',"NO₂",'PM10','PM2.5']
    years = [2015,2016,2017,2018,2019]

    df = df.loc[(df['CO']>0)&(df['PM10']>0)&(df['PM2.5']>0)&(df['NO₂']>0)]
    for element in elements:
        print()
        for year in years:
            print(df[element].loc[(df['year']==year) & (df[element]>0)].mean()) # 평균
            print(df[element].loc[(df['year']==year) & (df[element]>0)].std())  # 표준편차
            print(df[element].loc[(df['year']==year) & (df[element]>0)].min())  # 최소값
            print(df[element].loc[(df['year']==year) & (df[element]>0)].max())  # 최대값
            
#             print(np.round_(df[element].loc[(df['year']==year) & (df[element]>0)].max(),1))  # 최대값

            
            
# NO2는 1000배 해주어야 한다. 어떻게 처리할지 생각해볼것
# 이거 해결하면 엑셀에 복붙 뚝딱 끝
                  
```

```python
solution(df)
```

# 월별로 묶어줌 awesome

```python
year2 = df['time'].dt.year
month2 = df['time'].dt.month
result = df.groupby(month2).mean()
result.index = pd.date_range('1/1/2019', '12/1/2019', freq='MS').strftime('%B')
result

datetime.datetime.combine(year2, month2)
```
